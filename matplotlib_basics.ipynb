
### 1. line plot

```python
import matplotlib.pyplot as plt
# 添加中文显示：from matplotlib import rcParams
# rcParams['font.family']='SimHei'
# create the figure and set size
plt.figure(figsize=(10,5))

# data
month=['Jan','Feb','Mar','Apr']
sales=[100,150,80,130]

# draw
plt.plot(month,sales,label='product a',linewidth=2,marker='o')

#title
plt.title('sales trend',color='blue',fontsize=20)

# axis labels
plt.xlabel('month',fontsize=10)
plt.ylabel('sales($)',fontsize=10)

# legend
plt.legend(loc='upper left')

# add grid
plt.grid(True,alpha=0.5, color='red',linestyle='--')
# plt.grid(axis='y')

# set the size of ticks
plt.xticks(rotation=60,fontsize=20)

# set y axis range
plt.ylim(0,160)

# show data point
for x,y in zip(month,sales):
    plt.text(x,y+1,str(y),ha='center',va='bottom',fontsize=10)
# display
plt.show()
```


### 2. Bar chart (column chart)

```python
import matplotlib.pyplot as plt
# 添加中文显示：from matplotlib import rcParams
# rcParams['font.family']='SimHei'
# create the figure and set size
plt.figure(figsize=(10,5))

# data
subjects=['chinese','math','english','science']
scores=[85,92,78,88]

# draw
plt.bar(subjects,scores,label='student',color='orange',width=0.4)

# title
plt.title('student 2025 score record',color='red',fontsize=20)

# axis labels
plt.xlabel('subject',fontsize=10)
plt.ylabel('score',fontsize=10)

# legend
plt.legend(loc='upper right')

# add grid
plt.grid(axis='y',alpha=0.1, color='blue',linestyle='--')

# set the size of ticks
plt.xticks(rotation=0,fontsize=20)
plt.yticks(rotation=0,fontsize=20)

# set y axis range
plt.ylim(0,100)

# display data point
for x,y in zip(subjects,scores):
    plt.text(x,y+1,str(y),ha='center',va='bottom',fontsize=10)

# auto layout optimization
plt.tight_layout()
# display
plt.show()
```


### 3. Bar chart


### 4. Pie plot


```python
import matplotlib.pyplot as plt

# 添加中文显示：from matplotlib import rcParams
# rcParams['font.family']='SimHei'
# create the figure and set size
plt.figure(figsize=(10, 5))

# data
activities=['study','leisure','sport','sleep','other']
time=[6,4,1,8,5]
colors=['#66b3ff','#99ff99','#ffcc99','#ff9999','#ff4499']

# draw
plt.pie(time,
        labels=activities,
        autopct='%1.1f%%',
        startangle=90,
        colors=colors)

# title
plt.title('Daily activities', color='red', fontsize=20)

# auto layout optimization
plt.tight_layout()
# display
plt.show()
```


### 5. Ring plot

```python

### 3. Bar chart
import matplotlib.pyplot as plt

# 添加中文显示：from matplotlib import rcParams
# rcParams['font.family']='SimHei'
# create the figure and set size
plt.figure(figsize=(10, 5))

# data
activities=['study','leisure','sport','sleep','other']
time=[6,4,1,8,5]
colors=['#66b3ff','#99ff99','#ffcc99','#ff9999','#ff4499']

# draw
plt.pie(time,
        labels=activities,
        autopct='%1.1f%%',
        startangle=90,
        colors=colors,
        wedgeprops={'width':0.5},  # set the width of th ring
        pctdistance=0.8)           # set the distance of the percentage from the center

# title
plt.title('Daily activities', color='red', fontsize=20)
plt.text(0,0,'sum:100%',ha='center',fontsize=10)
# auto layout optimization
plt.tight_layout()
# display
plt.show()
```


### 6. Explode pie chart


```python
import matplotlib.pyplot as plt

# 添加中文显示：from matplotlib import rcParams
# rcParams['font.family']='SimHei'
# create the figure and set size
plt.figure(figsize=(10, 5))

# data
activities=['study','leisure','sport','sleep','other']
time=[6,4,1,8,5]
colors=['#66b3ff','#99ff99','#ffcc99','#ff9999','#ff4499']
explode=[0.1,0,0,0,0]
# draw
plt.pie(time,
        labels=activities,
        autopct='%1.1f%%',
        startangle=90,
        colors=colors,
        explode=explode,
        shadow=True
        )
#
# title
plt.title('Daily activities', color='red', fontsize=20)

# auto layout optimization
plt.tight_layout()
# display
plt.show()
```


### 7. Scatter plot

```python
import matplotlib.pyplot as plt
import random
# 添加中文显示：from matplotlib import rcParams
# rcParams['font.family']='SimHei'
# create the figure and set size
plt.figure(figsize=(10,7))

# data
x=[]
y=[]
for i in range(1000):
    tmp=random.uniform(0,10)
    x.append(tmp)
    tmp2=2*tmp+random.gauss(0,2)
    y.append(tmp2)

# draw
plt.scatter(x,y,
            color='blue',
            alpha=0.5,
            s=20)    # scatter(x,y)

#title
plt.title('relation of x and y',color='red',fontsize=20)

# axis labels
plt.xlabel('x',fontsize=10)
plt.ylabel('y',fontsize=10)


# add grid
plt.grid(True,alpha=0.5,linestyle='--')
# plt.grid(axis='y')

# set the size of ticks
plt.xticks(rotation=0,fontsize=15)
plt.yticks(rotation=0,fontsize=15)

# set y axis range
plt.ylim(0,30)
plt.plot([0,10],[0,20],color='orange',linewidth=2)  # regression line

# display
plt.show()
```


### 8. Box plot

```python
import matplotlib.pyplot as plt

data={
    'chinese':[82,85,88,70,90,76,84,83,95],
    'math':[75,80,79,93,88,82,87,89,92],
    'english':[70,72,68,65,78,80,85,90,95]
}
plt.figure(figsize=(8,6))
plt.boxplot(data.values(),tick_labels=data.keys())

plt.title("box plot of scores")
plt.ylabel("score")
plt.grid(axis='y',linestyle='--',alpha=0.5)
plt.show()
```

### sample question: Temperature analysis

```python
# 1. import library
import matplotlib.pyplot as plt
import pandas as pd

# 2. import data
df=pd.read_csv('data/weather.csv')
print(df.head())
print('-'*30)
```

```python
# 3. sketch temperature trend
df['date']=pd.to_datetime((df['date']))
df=df[df['date'].dt.year==2015]
plt.figure(figsize=(15,10))
plt.plot(df['date'],df['temp_max'],label='maximum temperature')
plt.plot(df['date'],df['temp_min'],label='minimum temperature')
plt.title('temperature variation in 2015')
plt.xlabel('date')
plt.ylabel('temperature')
plt.legend()
```

```python
df['temp_mean']=(df['temp_max']+df['temp_min'])/2
df.head()
```

```python
plt.figure(figsize=(15,10))
plt.plot(df['date'],df['temp_mean'],label='mean temperature')
plt.title('temperature variation in 2015')
plt.xlabel('date')
plt.ylabel('temperature')
plt.legend()
```
